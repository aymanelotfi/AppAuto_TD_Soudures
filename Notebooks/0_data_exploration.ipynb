## Missing Data

```python
filterd_weld_df = weld_df
nb_instances, nb_features = filterd_weld_df.shape
nan_values_count = filterd_weld_df.isnull().sum().sort_values()
nan_values_percent = nan_values_count/nb_instances*100

display(nan_values_percent.map('{:.2f} %'.format))
```

```python
plot = nan_values_percent.plot(kind="barh", title="Missing Values Plot (%)", figsize=(8,8), grid=True)
```

```python
# Calculate NaN fraction for each weld_type
na_fraction = weld_df.isna().groupby(weld_df["weld_type"]).mean()

# Sort the rows based on the number of instances in each weld_type
weld_type_counts = weld_df["weld_type"].value_counts()
na_fraction = na_fraction.reindex(weld_type_counts.index)

# Sort the columns based on the NaN fraction for the first weld_type
sorted_columns = na_fraction.loc[na_fraction.index[0]].sort_values().index
na_fraction = na_fraction.reindex(sorted_columns, axis=1)

# Create heatmap
plt.figure(figsize=(16, 10))
sns.heatmap(
    na_fraction,
    annot=True,
    cmap="OrRd",
    cbar_kws={"label": "Fraction of NaN values"},
)
plt.title("Heatmap of NaN Fractions by weld_type")
plt.ylabel("Weld Type")
plt.xlabel("Columns")
plt.tight_layout()
plt.show()
```

## Material composition features


```python
na = weld_df[material_composition_variables].isna().mean()

numeric = (
    weld_df[material_composition_variables]
    .apply(pd.to_numeric, errors="coerce")
    .notna()
    .mean()
)


na_numeric_ratio_df = pd.DataFrame(
    {
        "NA": na,
        "Numeric": numeric,
    }
)

na_numeric_ratio_df.sort_values("NA", inplace=True)

na_numeric_ratio_df["Non-numeric"] = (
    1 - na_numeric_ratio_df["NA"] - na_numeric_ratio_df["Numeric"]
)

na_numeric_ratio_df.drop("NA", axis=1, inplace=True)

ax = na_numeric_ratio_df.plot(
    kind="bar",
    stacked=True,
    figsize=(11, 7),
    xlabel="Material composition variables",
    ylabel="Percentage of instances with values reported",
    rot=45,
)
ax.yaxis.set_major_formatter(PercentFormatter(1))
ax.set_ylim(0, 1)
```

```python
for variable in welding_process_variables:
    print("-" * 80)
    print(variable)
    print(f"Percentage of NA values: {weld_df[variable].isna().mean():.1%}")
    print("Unique values:", *weld_df[variable].unique())
    print(
        weld_df[variable][pd.to_numeric(weld_df[variable], errors="coerce").isna()]
        .value_counts(dropna=False)
        .rename("Counts for non-numeric values")
    )
```
